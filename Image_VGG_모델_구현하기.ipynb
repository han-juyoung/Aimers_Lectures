```python
import tensorflow as tf
from tensorflow.keras import Sequential, layers

def build_vgg16():
    # Sequential 모델 선언
    model = Sequential()

    # TODO: [지시시항 1번] 첫번째 Block을 완성하세요.
    model.add(layers.Conv2D(64, kernel_size=(3,3), padding="same", activation="relu", input_shape=(224, 224, 3))) # 1개
    model.add(layers.Conv2D(64, kernel_size=(3,3), padding="same", activation="relu")) # 2개
    model.add(layers.MaxPooling2D(2))

    # TODO: [지시시항 2번] 두번째 Block을 완성하세요.
    model.add(layers.Conv2D(128, kernel_size=(3,3), padding="same", activation="relu")) # 3개
    model.add(layers.Conv2D(128, kernel_size=(3,3), padding="same", activation="relu")) # 4개
    model.add(layers.MaxPooling2D(2))

    # TODO: [지시시항 3번] 세번째 Block을 완성하세요.
    model.add(layers.Conv2D(256, kernel_size=(3,3), padding="same", activation="relu")) # 5개
    model.add(layers.Conv2D(256, kernel_size=(3,3), padding="same", activation="relu")) # 6개
    model.add(layers.Conv2D(256, kernel_size=(3,3), padding="same", activation="relu")) # 7개
    model.add(layers.MaxPooling2D(2))

    # TODO: [지시시항 4번] 네번째 Block을 완성하세요.
    model.add(layers.Conv2D(512, kernel_size=(3,3), padding="same", activation="relu")) # 8개
    model.add(layers.Conv2D(512, kernel_size=(3,3), padding="same", activation="relu")) # 9개
    model.add(layers.Conv2D(512, kernel_size=(3,3), padding="same", activation="relu")) # 10개
    model.add(layers.MaxPooling2D(2))

    # TODO: [지시시항 5번] 다섯번째 Block을 완성하세요.
    model.add(layers.Conv2D(512, kernel_size=(3,3), padding="same", activation="relu")) # 11개
    model.add(layers.Conv2D(512, kernel_size=(3,3), padding="same", activation="relu")) # 12개
    model.add(layers.Conv2D(512, kernel_size=(3,3), padding="same", activation="relu")) # 13개
    model.add(layers.MaxPooling2D(2))

    # Fully Connected Layer
    model.add(layers.Flatten())
    model.add(layers.Dense(4096, activation="relu")) # 14개
    model.add(layers.Dense(4096, activation="relu")) # 15개
    model.add(layers.Dense(1000, activation="softmax")) # 16개

    # 레이어 개수 셀 때 파라미터 포함된 레이어만 센다.
    # Flatten 이나 MaxPool2D 같은 단순 작업은 포함되지 않는다.

    return model

def main():
    model = build_vgg16()
    model.summary()

if __name__ == "__main__":
    main()
```
