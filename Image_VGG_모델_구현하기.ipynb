```python
import tensorflow as tf
from tensorflow.keras import Sequential, layers

def build_vgg16():
    # Sequential  
    model = Sequential()

    # TODO: [ 1]  Block .
    model.add(layers.Conv2D(64, kernel_size=(3,3), padding="same", activation="relu", input_shape=(224, 224, 3))) # 1
    model.add(layers.Conv2D(64, kernel_size=(3,3), padding="same", activation="relu")) # 2
    model.add(layers.MaxPooling2D(2))

    # TODO: [ 2]  Block .
    model.add(layers.Conv2D(128, kernel_size=(3,3), padding="same", activation="relu")) # 3
    model.add(layers.Conv2D(128, kernel_size=(3,3), padding="same", activation="relu")) # 4
    model.add(layers.MaxPooling2D(2))

    # TODO: [ 3]  Block .
    model.add(layers.Conv2D(256, kernel_size=(3,3), padding="same", activation="relu")) # 5
    model.add(layers.Conv2D(256, kernel_size=(3,3), padding="same", activation="relu")) # 6
    model.add(layers.Conv2D(256, kernel_size=(3,3), padding="same", activation="relu")) # 7
    model.add(layers.MaxPooling2D(2))

    # TODO: [ 4]  Block .
    model.add(layers.Conv2D(512, kernel_size=(3,3), padding="same", activation="relu")) # 8
    model.add(layers.Conv2D(512, kernel_size=(3,3), padding="same", activation="relu")) # 9
    model.add(layers.Conv2D(512, kernel_size=(3,3), padding="same", activation="relu")) # 10
    model.add(layers.MaxPooling2D(2))

    # TODO: [ 5]  Block .
    model.add(layers.Conv2D(512, kernel_size=(3,3), padding="same", activation="relu")) # 11
    model.add(layers.Conv2D(512, kernel_size=(3,3), padding="same", activation="relu")) # 12
    model.add(layers.Conv2D(512, kernel_size=(3,3), padding="same", activation="relu")) # 13
    model.add(layers.MaxPooling2D(2))

    # Fully Connected Layer
    model.add(layers.Flatten())
    model.add(layers.Dense(4096, activation="relu")) # 14
    model.add(layers.Dense(4096, activation="relu")) # 15
    model.add(layers.Dense(1000, activation="softmax")) # 16

    #        .
    # Flatten  MaxPool2D     .

    return model

def main():
    model = build_vgg16()
    model.summary()

if __name__ == "__main__":
    main()
```
