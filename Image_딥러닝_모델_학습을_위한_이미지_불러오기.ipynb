```python
from elice_utils import EliceUtils

elice_utils = EliceUtils()

from PIL import Image
import matplotlib.pyplot as plt
import os
import numpy as np

def show_plot(img, title=" "):
    plt.title(title)
    plt.imshow(img)
    plt.savefig("tmp.png")
    elice_utils.send_image("tmp.png")

def load_image(path, name):
    # TODO: [ 1]    

    # path = "dataset/val/dogs"
    # name = "dog.0.jpg"
    # => "dataset/val/dogs/dog.0.jpg"
    # path + "/" + name  open    

    img = Image.open(os.path.join(path, name))
    return img

def main():
    data_path = "dataset/val/dogs"


    #   plt  
    names = os.listdir(data_path)
    img = load_image(data_path, names[0])

    #   
    show_plot(img, "PIL original image")


    # TODO: [ 2]       .
    # PIL    
    pil_size = img.size
    print("PIL   :", pil_size)
    # PIL   : (499, 375) == (, )

    # PIL  numpy  
    np_img = np.array(img)

    # numpy  shape 
    np_shape = np_img.shape
    print("numpy  shape:", np_shape)
    # numpy  shape: (375, 499, 3) == (, , )
    show_plot(np_img, "Numpy array image")

    # TODO: [ 3] PIL numpy     .
    # PIL.Image x=10, y=20   
    pil_pix = img.load()[10, 20] # , 

    # numpy  x=10, y=20   
    np_pix = np_img[20, 10] # (), ()
    print("PIL : {}, numpy : {}".format(pil_pix, np_pix))

    # PIL    (224,224) .
    resized_img = img.resize((224, 224)) # ImageNet 

    # resize  
    show_plot(resized_img, "Resized image")
    print("resize  :", resized_img.size)

    return pil_size, np_img, np_shape, np_pix, resized_img

if __name__ == "__main__":
    main()
```
