```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings(action='ignore') #   

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix #  
"""
1. data    dataset.csv ,
       
     .

   Step01. pandas read_csv()  
           data    dataset.csv
           .

   Step02.  X y .
              dataset.csv 
           ,

           X  'Class'    ,
           y  'Class'   .
"""
def load_data():

    data = pd.read_csv('data/dataset.csv')

    X = data.drop('Class', axis=1) # default =  (axis=0)
    y = data['Class']

    train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.2, random_state = 0)
    print(X, y) #      , print(data)
    return train_X, test_X, train_y, test_y

"""
2. SVM  ,
       ,
         
   .

   Step01. SVM  .

   Step02. SVM    
           .

   Step03.   
               .
"""
def SVM(train_X, test_X, train_y, test_y):

    svm = SVC() # SV Classifier

    svm.fit(train_X, train_y)

    pred_y = svm.predict(test_X)

    return pred_y

#  ,     main .
def main():

    train_X, test_X, train_y, test_y = load_data()

    pred_y = SVM(train_X, test_X, train_y, test_y)

    # SVM   .
    print("\nConfusion matrix : \n",confusion_matrix(test_y,pred_y)) # 
    print("\nReport : \n",classification_report(test_y,pred_y))

if __name__ == "__main__":
    main()
```
