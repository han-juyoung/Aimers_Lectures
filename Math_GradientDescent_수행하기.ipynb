```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(100)
y = 0.7 * x
```

```python
# 문제 시각화. 주어진 x 에 대응되는 y 예측하기

plt.figure(figsize=(8,6))
plt.scatter(x,y)
plt.show()
```

```python
# Y = Wx + b 로 예상하여, W와 b를 gradient descent를 통해 찾는다
W = np.random.uniform(0, 1) # 초기값을 랜덤하게 설정
b = np.random.uniform(0, 1)

step_size = 0.5 # 이동거리, learning rate

for epoch in range(100):
  y_pred = W * x + b # 출력값(예측값)

  objective = np.abs(y_pred - y).mean() # 목적함수 = (예측값 - 정답) 평균을 냄. 최소가 되도록 하는게 목적. 그것을 미분을 한다.

  w_grad = step_size * ((y_pred-y)*x).mean()
  b_grad = step_size * (y_pred-y).mean()

  W = W - w_grad
  b = b - b_grad
```

```python
# 학습 전 후의 예측함수의 변화 과정 시각화
def compare_pred(x, pred, y):
  plt.figure(figsize=(8,6))
  plt.scatter(x,y, label='ground-truth')
  plt.scatter(x,pred,label='prediction')
  plt.legend()
  plt.show()

pred = W * x  + b
compare_pred(x, pred, y)
```
