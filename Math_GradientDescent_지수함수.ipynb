```python
from elice_utils import EliceUtils
import numpy as np
import matplotlib.pyplot as plt

elice_utils = EliceUtils()



def main():

    gradient()

def gradient():
    '''
     :   f   df .
    '''
    cur_x = 3 #  x
    lr = 0.01 # Learning rate
    threshold = 0.000001 #  while  .
    previous_step_size = 1 #
    max_iters = 10000 #  iteration 
    iters = 0 #iteration counter
    f = lambda x: np.exp(x+1) #    
    df = lambda x: np.exp(x+1) #   

    history_x = []
    history_y = []
    while previous_step_size > threshold and iters < max_iters:
        prev_x = cur_x #  x prev .
        cur_x = cur_x - lr * df(prev_x) #Grad descent
        previous_step_size = abs(cur_x - prev_x) #x  .
        iters = iters+1 #iteration count
        print("Iteration",iters,"\nx  ",cur_x, "\ny  ", f(cur_x)) #Print iterations
        history_x.append(cur_x)
        history_y.append(f(cur_x))

    plt.plot(history_x, history_y, marker='o')
    plt.show()

    #     
    plt.savefig("history.png")
    elice_utils.send_image("history.png")

if __name__ == "__main__":
    main()
```
