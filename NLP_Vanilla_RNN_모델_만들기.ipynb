```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
#   

import tensorflow as tf
from tensorflow.keras import layers, Sequential

# TODO: [ 1]   .
def build_model1():
    model = Sequential()

    model.add(layers.Embedding(10, 5)) # (  , )  (input, output)  
    model.add(layers.SimpleRNN(3)) # hidden state      

    return model

# TODO: [ 2]   .
def build_model2():
    model = Sequential()

    model.add(layers.Embedding(256, 100))
    model.add(layers.SimpleRNN(20))
    model.add(layers.Dense(10, activation="softmax")) #   ,       

    return model

def main():
    model1 = build_model1()
    print("=" * 20, " ", "=" * 20)
    model1.summary()

    print()

    model2 = build_model2()
    print("=" * 20, " ", "=" * 20)
    model2.summary()

if __name__ == "__main__":
    main()
```
